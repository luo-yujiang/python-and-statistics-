```python
import pandas as pd
df1 = pd.read_excel('data2.xlsx')
print(df1)
z=df1.iloc[:,1:]
region=df1.iloc[:,0]
```

```python
titles = df1.columns
titles = list(titles)
d=dict()
for j in range(6):
    d[j]=(titles[j])
```

```python
print('最小值\n',df1.min(),end='\n\n')
print('最大值\n',df1.max(),end='\n\n')
print('标准差\n',df1.std(),end='\n\n')
print('方差\n',df1.var(),end='\n\n')
```

```python
print('均值\n',df1.mean(),end="\n\n")
print('四分位数\n',df1.quantile(q=0.25),end='\n\n')
print('峰度\n',df1.skew(),end='\n\n')
print('偏度\n',df1.kurt(),end='\n\n')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
```

```python
# bar图
for i in range(6):
    fig = plt.figure()
    fig.set_figheight(20)
    fig.set_figwidth(50)
    plt.subplot(3,2,1+i)
    a = z.iloc[:,i]
    b = a.values.flatten()
    Re = region.values.flatten()
    plt.title(titles[i+1])
    plt.bar(Re,b)
    plt.show()
```

```python
# box图
for i in range(6):
    fig = plt.figure()
    fig.set_figheight(20)
    fig.set_figwidth(20)
    plt.title(titles[i+1])
    plt.subplot(3,2,1+i)
    sns.boxplot(data=z.iloc[:,i])
```

```python
for i in range(6):
    fig = plt.figure()
    fig.set_figheight(20)
    fig.set_figwidth(50)
    plt.subplot(3,2,1+i)
    a = z.iloc[:,i]
    b = a.values.flatten()
    Re = region.values.flatten()
    plt.title(titles[i+1])
    plt.pie(b,labels=Re)
    plt.show()
```
